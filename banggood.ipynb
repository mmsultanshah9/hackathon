# 2. Data Cleaning and Transformation

```python
import pandas as pd
import numpy as np

# Load CSV
df = pd.read_csv("/mnt/data/banggood_scraped_products.csv")

# ---- CLEAN PRICE AND REPLACE IN SAME COLUMN ----
def extract_price(x):
    if pd.isna(x):
        return np.nan
    x = str(x).replace(",", "")
    nums = ''.join(ch if ch.isdigit() or ch == "." else "" for ch in x)
    return float(nums) if nums else np.nan

df["price"] = df["price"].apply(extract_price)

# ---- CLEAN RATING AND REPLACE IN SAME COLUMN ----
def extract_rating(x):
    if pd.isna(x):
        return np.nan
    nums = ''.join(ch if ch.isdigit() or ch == "." else "" for ch in str(x))
    return float(nums) if nums else np.nan

df["rating"] = df["rating"].apply(extract_rating)

# ---- CLEAN REVIEWS AND REPLACE IN SAME COLUMN ----
def extract_reviews(x):
    if pd.isna(x):
        return np.nan
    nums = ''.join(ch for ch in str(x) if ch.isdigit())
    return int(nums) if nums else np.nan

df["reviews"] = df["reviews"].apply(extract_reviews)

# ---- HANDLE MISSING VALUES ----
df["price"].fillna(df["price"].median(), inplace=True)
df["rating"].fillna(df["rating"].mean(), inplace=True)
df["reviews"].fillna(0, inplace=True)

# ---- DERIVED FEATURE 1: PRICE IN PKR ----
df["price_pkr"] = df["price"] * 280   # 1 USD = 280 PKR

# ---- DERIVED FEATURE 2: RATING CATEGORY ----
df["rating_category"] = pd.cut(
    df["rating"],
    bins=[0, 2, 4, 5],
    labels=["Low", "Medium", "High"],
    include_lowest=True
)

# Show output
print(df.head())

# Save cleaned file
df.to_csv("cleaned_replaced_banggood.csv", index=False)
print("Saved: cleaned_replaced_banggood_products.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# --- 1. Data Loading and Preparation ---

# Load the dataset
df = pd.read_csv('cleaned_replaced_banggood.csv')

# Extract the main category from the category_url (e.g., 'Wholesale-Smart-Remote-Control-c-...')
# This is necessary for analyses involving 'per category' metrics.
df['main_category'] = df['category_url'].apply(
    lambda x: re.search(r'Wholesale-(.*?)-c-\d+', x).group(1)
    if re.search(r'Wholesale-(.*?)-c-\d+', x) else 'Other'
)

# Calculate 'Value Metric' for Analysis 4: Rating / Log(1 + Price_PKR)
# A higher score indicates better value (high rating at a low price).
df['value_metric'] = df['rating'] / np.log1p(df['price_pkr'])

print("--- Data Preparation Complete ---")
print(f"Unique Categories found: {df['main_category'].unique()}\n")


# --- 2. Exploratory Analyses ---

# Analysis 1: Price Distribution per Category
plt.figure(figsize=(8, 6))
# Use log-transformed price for better visualization of the heavily skewed distribution
sns.boxplot(x='main_category', y=np.log1p(df['price_pkr']), data=df, palette='pastel')
plt.title('1. Log-Transformed Price Distribution per Main Category')
plt.xlabel('Main Category')
plt.ylabel('Log(1 + Price in PKR)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('1_price_distribution_per_category.png')
plt.close()
print("1. Plot for Price Distribution per Category saved as '1_price_distribution_per_category.png'")


# Analysis 2: Rating vs Price Correlation (Scatter Plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='rating', y=np.log1p(df['price_pkr']), data=df, alpha=0.6)
plt.title('2. Rating vs. Log-Transformed Price (PKR)')
plt.xlabel('Rating')
plt.ylabel('Log(1 + Price in PKR)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig('2_rating_vs_log_price_correlation.png')
plt.close()
print("2. Plot for Rating vs Price Correlation saved as '2_rating_vs_log_price_correlation.png'")


# --- Data Preparation ---
# Extract the main category
df['main_category'] = df['category_url'].apply(
    lambda x: re.search(r'Wholesale-(.*?)-c-\d+', x).group(1)
    if re.search(r'Wholesale-(.*?)-c-\d+', x) else 'Other'
)
# Calculate 'Value Metric'
df['value_metric'] = df['rating'] / np.log1p(df['price_pkr'])


# --- Figure 1: Top 5 Reviewed Products ---
top_reviewed = df.sort_values(by='reviews', ascending=False).head(5)

plt.figure(figsize=(10, 6))
# Use horizontal bar chart for better readability of product names
sns.barplot(x='reviews', y='product_name', data=top_reviewed, palette='magma')
plt.title('Top 5 Products by Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Product Name')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('top_5_reviewed_products_figure.png')
plt.close()


# --- Figure 2: Best Value Metric per Category ---
best_value_products = df.loc[df.groupby('main_category')['value_metric'].idxmax()]
# Sort by value_metric for clarity
best_value_products = best_value_products.sort_values(by='value_metric', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x='main_category', y='value_metric', data=best_value_products, palette='viridis')
plt.title('Best Value Metric per Category')
plt.xlabel('Main Category')
plt.ylabel('Value Metric (Rating / Log(1 + Price))')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('best_value_metric_per_category_figure.png')
plt.close()

print("Figures generated for Top Reviewed Products and Best Value Metric per Category.")


# Analysis 5 (Substitution): Correlation Matrix of Numerical Features
numerical_df = df[['price_pkr', 'rating', 'reviews']]
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(7, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('5. Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.savefig('5_correlation_matrix.png')
plt.close()
print("\n5. Plot for Correlation Matrix saved as '5_correlation_matrix.png'")
```
